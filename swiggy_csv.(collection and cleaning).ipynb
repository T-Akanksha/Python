Restaurant,Cuisine,Coupon,Rating,Delivered_in,Budget,People


```python
restaurant=[]
cuisine=[]
coupon=[]
rating=[]
delivered_in=[]
budget=[]
people=[]
for i in range(1,9):
    url="https://www.swiggy.com/city/bangalore/south-indian-collection?page="+str(i)
    print(url)
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    a= soup.find_all('div', class_="nA6kb")
    for i in a:
        restaurant.append(i.text)
        
    b= soup.find_all('div', class_="_1gURR")
    for i in b:
        cuisine.append(i.text)  
        
    d= soup.find_all('span', class_="sNAfh")
    for i in d:
        coupon.append(i.text)
        
    
    e=soup.find_all('div', class_="_3Mn31")
    for i in e:
        rating.append(i.text.split("•")[0])
        delivered_in.append(i.text.split("•")[1].split(" ")[0])
        budget.append(i.text.split("₹")[1].split(" ")[0])
        
        
print(len(restaurant))
print(len(cuisine))
print(len(coupon))
print(len(rating))
print(len(delivered_in))
print(len(budget))
```

```python
df=pd.DataFrame({'Restaurant':restaurant, 'Cuisine':cuisine, 'Coupon':coupon[0:128], 'Rating':rating,"Delivery_time":delivered_in,"Price per two":budget})
```

```python
city=["Bangalore"]
r=128
l= []
for i in city:
    for ele in range(r):
        l.append(i)
print(l)
```

```python
df["Location"]=l
```
