Univarient:
---

The simplest type of analysis, known as a univariate analysis, focuses on just one variable. 

Performing Univariate Analysis for Numerical and Categorical data.

**1. CountPlot**

Countplot is basically a count of frequency plot in form of a bar graph.It plots the count of each category in a separate bar.


let us plot a countplot for categorical data.

```python
sns.countplot(swiggy['Location'])
plt.show()
```

- The countplot above represents the number of restaurants in the respective locations. Maximum restaurants are in hyderabad and minimum are in chennai

```python
plt.figure(figsize=(20,8))
sns.countplot(swiggy['Coupon'])
plt.xticks(rotation=90)
plt.show()
```

- Coupon attribute count plot shows more restaurants have coupons 50% off|Use WELCOME 50

**2.Distplot**

A Distplot or distribution plot, depicts the variation in the data distribution.Distplot lets you show a histogram with a line on it. 

Distplot represents the overall distribution of continuous data variables.

It explains PDF(Probability Density Function) which means what is the probability of each value occurring in this column. 

```python
sns.distplot(swiggy["Rating"])
plt.show()
```

- The area under the density curve represents the theoretical probability of that column.The more restuarents have the rating in middle of 3.8 to 4.3 

```python
sns.distplot(swiggy["Delivery_time"])
plt.show()
```

- The distplot above shows the more restuareants have the delivery time in between 30 to 40 minutes. 

```python
sns.distplot(swiggy["Price per two"])
plt.show()
```

- The above displot represents the Price per two people is started from 150 to 250 rupees.

**3.Pie Chart**
- The pie chart is also the same as the countplot, only gives you additional information about the percentage presence of each category in data means which category is getting how much weightage in data. 
- let us check about the Location column, what is the percentage of restuarants in located in that location.

```python
swiggy["Location"].value_counts().plot(kind="pie",autopct="%.2f",radius=1)
plt.show()
```

- The pie chart above shows the percentage of restaurants in the locations.The highest percentage of restuarants are in Hyderabad and the lowest percentage of restuarants are in Chennai.

**4. BoxPlot**

Boxplot is a method for graphically depicting groups of numerical data through their quartiles.


The first quartile is the point at which 25% of the data is below, while the third quartile represents the point at which 75% of the data is above. The median is the data point line in the centre, with half of the points above and half below this number.

let us plot boxplot for Numerical data.

```python
swiggy.boxplot(column=["Rating"])
plt.show()
```

- The lowest outlier is starting from 1.8 to 3.2 and the extreme outlier is only one lying between 4.5 and 5.0.

```python
swiggy.boxplot(column=["Delivery_time"])
plt.show()
```

- The box chart above shows an median delivery time of 35 minutes. The lines that extend above and below the box are called whiskers. These values connect the box to the maximum and minimum values that are within 1.5 times the IQR range of the box. The dots show data points that lie outside the whiskers, which are generally considered outliers.

- The extreme outliers of ddelivery time are starting from 55 and above 60 minutes.

```python
swiggy.boxplot(column=["Price per two"])
plt.show()
```

- The box chart above shows that the median value of Price per two column is 250.The extreme outlier values for price per two are from 600 upto 1200.

Bivarient:
----

Bivariate analysis is one of the statistical analysis where two variables are observed. One variable here is dependent while the other is independent. 

Our main task is to explore the relationship between variables to build a powerful.

**1. Numerical and Numerical**
- First, let’s explore the plots when both the variable is numerical.

**a. Scatter Plot**
- To plot the relationship between two numerical variables scatter plot is a simple plot to do. 

```python
sns.scatterplot(swiggy["Rating"], swiggy["Price per two"])
plt.show()
```

- Restaurants with a maximum price of 1200 for two people have a rating of 4.4, and restaurants with a minimum rating of 1.5 have a price of 250 for two people.

```python
sns.scatterplot(swiggy["Delivery_time"], swiggy["Price per two"])
plt.show()
```

- Restaurants with a maximum price of 1200 have a delivery time of more than 50 minutes, and a restaurant with a minimum price of 0 to 50 rupees has a delivery time of 35 minutes.

**2. Numerical and Categorical**
- If one variable is numerical and one is categorical then there are various plots that we can use for Bivariate and Multivariate analysis.

**a. Bar Plot**
- Bar plot is a simple plot which we can use to plot categorical variable on the x-axis and numerical variable on y-axis and explore the relationship between both variables. 

- The blacktip on top of each bar shows the confidence Interval.

```python
sns.barplot(swiggy['Location'], swiggy['Price per two'])
plt.show()
```

- The barplot above shows that the highest price for the two people is Noida and the lowest price for two people is Lucknow.

```python
sns.barplot(swiggy['Location'], swiggy['Rating'])
plt.show()
```

- Restaurants in the Banglore location have a maximum rating and the Lucknow restaurants has a minimum rating compared to other locations.

```python
plt.figure(figsize=(20,8))
sns.barplot(swiggy['Coupon'], swiggy['Price per two'])
plt.xticks(rotation=90)
plt.show()
```

- Promo code offer minimum 20% off|TRYNEW is applied to budgets with less than 100 and maximum promo code offer of 40%|TRYNEW is applied to budgets with 650.

Multivarient:
----

When we analyze more than 2 variables together then it is known as Multivariate Analysis. we will work on different plots for Bivariate as well on Multivariate Analysis.

**a. Multivariate analysis with scatter plot**
- we can also plot 3 variable or 4 variable relationships with scatter plot. 
- suppose we want to find the separate ratio Location with Rating and Price per two provided.
- And also want to find the separate ratio Location with Price per two and Delivery time provided.

```python
sns.scatterplot(swiggy["Rating"], swiggy["Price per two"],hue=swiggy["Location"])
plt.show()
```

- Hyderabad has a maximum price of 1200 for two people have a rating of 4.4, and Lucknow has a minimum rating of 1.5 have a price of 250 for two people.

```python
sns.scatterplot(swiggy["Price per two"], swiggy["Delivery_time"],hue=swiggy["Location"])
plt.show()
```

- Hyderabad has a maximum price of 1200 have a delivery time of more than 50 minutes, and kochi has a minimum price of 0 to 50 rupees has a delivery time of 35 minutes.

**b. Violin Plot**
- Violinplot Displays distribution of data across all levels of a category.

```python
sns.violinplot(data=swiggy,y="Rating",x="Location")
plt.show()
```

- There are fewer outliers in Chennai location and more outliers in lucknow location.

### **Heatmap:**

- The heatmap is a way of representing the data in a 2-dimensional form. The data values are represented as colors in the graph. The goal of the heatmap is to provide a colored visual summary of information.

- In Exploratory data analysis, a heatmap allows us to examine the correlation matrix (EDA). The linear relationship between two variables is determined by correlation. It not only tells us the strength of the relationship, but also the direction of the relationship, i.e., whether the variables are directly or inversely related.

- To add text over the heatmap, we can use the annot attribute. If annot is set to True, the text will be written on each cell.

```python
sns.heatmap(swiggy.corr(), annot=True)
plt.show()
```

- +1 represents the exact opposite,i.e., as one variable increases, the other increases in the same proportion.

- Other value between the range of -0.2 to 1 the low or high correlation.
